```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Reading Dataset

```python
df = pd.read_csv(r'.\weight-height.csv')
df.sample(3)
```

```python
male = df[df['Gender'] == 'Male']
female = df[df['Gender'] == 'Female']
```

### Creating plot

```python
grid = plt.GridSpec(5, 5, wspace=1, hspace=1)
ax1 = plt.subplot(grid[:3,:2])
ax2 = plt.subplot(grid[:3,2:])
ax3 = plt.subplot(grid[3:,2:])
```

```python
grid = plt.GridSpec(5,5, wspace=1, hspace=1)
ax1 = plt.subplot(grid[:3, :2])
ax2 = plt.subplot(grid[:3, 2:])
ax3 = plt.subplot(grid[3:, 2:])

ax1.hist(male['Height'], orientation='horizontal', alpha=.6)
ax1.hist(female['Height'], orientation='horizontal', alpha=.6)
```

```python
grid = plt.GridSpec(5,5, wspace=1, hspace=1)
ax1 = plt.subplot(grid[:3, :2])
ax2 = plt.subplot(grid[:3, 2:])
ax3 = plt.subplot(grid[3:, 2:])

ax1.hist(male['Height'], orientation='horizontal', alpha=.6)
ax1.hist(female['Height'], orientation='horizontal', alpha=.6)
ax1.invert_xaxis()
plt.show()
```

```python
grid = plt.GridSpec(5,5, wspace=1, hspace=1)
ax1 = plt.subplot(grid[:3, :2])
ax2 = plt.subplot(grid[:3, 2:])
ax3 = plt.subplot(grid[3:, 2:])

ax1.hist(male['Height'], orientation='horizontal', alpha=0.6)
ax1.hist(female['Height'], orientation='horizontal', alpha=0.6)
ax1.invert_xaxis()

ax2.scatter(male['Weight'], male['Height'], alpha=0.3)
ax2.scatter( female['Weight'], female['Height'], alpha=0.3)
```

```python
grid = plt.GridSpec(5,5, hspace=1, wspace=1)
ax1 = plt.subplot(grid[:3, :2])
ax2 = plt.subplot(grid[:3, 2:])
ax3 = plt.subplot(grid[3:, 2:])

ax1.hist(male['Height'], orientation='horizontal', alpha=0.6)
ax1.hist(female['Height'] , orientation='horizontal', alpha=0.6)
ax1.invert_xaxis()

ax2.scatter(male['Weight'], male['Height'], alpha=0.6)
ax2.scatter(female['Weight'], female['Height'], alpha=0.6)

ax3.hist(male['Weight'], alpha=0.6)
ax3.hist(female['Weight'], alpha=0.6)

plt.show()
```

```python
grid = plt.GridSpec(5,5, hspace=1, wspace=1)
ax1 = plt.subplot(grid[:3, :2])
ax2 = plt.subplot(grid[:3, 2:])
ax3 = plt.subplot(grid[3:, 2:])

ax1.hist(male['Height'], orientation='horizontal', alpha=0.6)
ax1.hist(female['Height'], orientation='horizontal', alpha=0.6)
ax1.invert_xaxis()

ax2.scatter(male['Weight'], male['Height'], alpha=0.6)
ax2.scatter(female['Weight'], female['Height'], alpha=0.6)

ax3.hist(male['Weight'], alpha=0.6)
ax3.hist(female['Weight'], alpha=0.6)
ax3.invert_yaxis()
```

```python
grid = plt.GridSpec(5,5, hspace=2, wspace=1)
ax1 = plt.subplot(grid[:3, :2])
ax2 = plt.subplot(grid[:3, 2:])
ax3 = plt.subplot(grid[3:, 2:])

ax1.hist(male['Height'], orientation='horizontal', alpha=0.5)
ax1.hist(female['Height'], orientation='horizontal', alpha=0.5)

ax2.scatter(male['Weight'], male['Height'], alpha=0.5, label='Male')
ax2.scatter(female['Weight'], female['Height'], alpha=0.5, label='Female')
ax2.legend(loc='upper left', bbox_to_anchor=(1,1))
ax2.set(xlabel='Weight', ylabel='Height')

ax3.hist(male['Height'], alpha=.5)
ax3.hist(female['Height'], alpha=.5)
ax3.invert_yaxis()


plt.suptitle('Height x Weight', size=20)
plt.show()
```
